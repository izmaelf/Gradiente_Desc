# Theory on the Gradient Descent algorithm

```python


# Definimos la función objetivo / función parabólica f(x, y) = x^2 + y^2

def func(x, y):
    return x**2 + y**2
# Definimos la gradiente para la función
def gradient(x, y):
    df_dx = 2 * x
    df_dy = 2 * y
    return np.array([df_dx, df_dy])

# Implementar el algoritmo de gradiente descendente
def gradient_descent(starting_point, learning_rate, num_iterations):
    x, y = starting_point
    positions = []  
    for i in range(num_iterations):
        positions.append((x, y)) 
        grad = gradient(x, y)
        x -= learning_rate * grad[0]
        y -= learning_rate * grad[1]
    return np.array(positions)

# Parámetros del algoritmo
starting_point = np.array([3.0, 4.0])  # Punto de inicio
learning_rate = 0.1  # Tasa de aprendizaje
num_iterations = 50  # Número de iteraciones

# Ejecutar el descenso de gradiente
positions = gradient_descent(starting_point, learning_rate, num_iterations)

```

```python
# Crear la gráfica 2D con contornos
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = func(X, Y)

# Crear la figura de contorno
plt.figure(figsize=(8, 6))
plt.contour(X, Y, Z, levels=50, cmap='viridis')

# Graficar la trayectoria del descenso de gradiente
plt.plot(positions[:, 0], positions[:, 1], color='r', marker='o', label='Trayectoria')

# Etiquetas y título
plt.title('Trayectoria del gradiente descendente sobre la función 2D')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()

# Mostrar la gráfica
plt.show()
```
