# ¿En que consiste la metodologia de gradiente descendente?

La metodología de descenso por gradiente (o gradient descent) es un algoritmo de optimización utilizado para minimizar  funciones,especialmente en problemas de aprendizaje automático y redes neuronales. El objetivo principal es encontrar los valores de los parámetros que minimicen una función de costo o pérdida. Con esto logra optimizar los procesos y costos en la investigación.

# Descenso de gradiente para una funcion 2D

El objetivo de esta practica para fuciones 2D es encontrar el numero minimo de iteraciones para poder tener con claridad las los valores optimizados gracias al numero de iteraciones que fueron necesarias para encontrar el minimo local.

```python
#Importamos las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

# Theory on the Gradient Descent algorithm

```python

# Definimos la función objetivo /Como ejemplo tomé una función parabólica: f(x, y) = x^2 + y^2

def func(x, y):
    return x**2 + y**2

# Definimos la gradiente para la función
def gradient(x, y):
    df_dx = 2 * x
    df_dy = 2 * y
    return np.array([df_dx, df_dy])

# Implementamos el algoritmo de gradiente descendente, donde estaremos almacenando el numero de iteraciones que se generen.
def gradient_descent(starting_point, learning_rate, num_iterations):
    x, y = starting_point
    positions = []  
    for i in range(num_iterations):
        positions.append((x, y)) 
        grad = gradient(x, y)
        x -= learning_rate * grad[0]
        y -= learning_rate * grad[1]
    return np.array(positions)

# Generamos los prámetros del algoritmo
starting_point = np.array([3.0, 4.0])  # Punto de inicio
learning_rate = 0.1  # Tasa de aprendizaje
num_iterations = 50  # Número de iteraciones

# Ejecutamos el descenso de gradiente.
positions = gradient_descent(starting_point, learning_rate, num_iterations)

```

```python
# Se crea la gráfica 2D con contornos para mejor visualización.
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = func(X, Y)

# Creamos la figura de contorno.
plt.figure(figsize=(8, 6))
plt.contour(X, Y, Z, levels=50, cmap='viridis')

# creamos la trayectoria del descenso de gradiente.
plt.plot(positions[:, 0], positions[:, 1], color='r', marker='o', label='Trayectoria')

# Etiquetas y título
plt.title('Trayectoria del gradiente descendente sobre la función (parabólica) 2D')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()

# Mostramos la gráfica
plt.show()
```
